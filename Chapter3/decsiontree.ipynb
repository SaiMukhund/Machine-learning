```python
## modules for data interpretation and visualization 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

## modules for training on the data 
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

#### Load the iris data

```python
iris=datasets.load_iris()
iris
```

#### Get the input and output variables for the data 

```python
X=iris.data
y=iris.target
columns=iris.feature_names
```

```python
X[:10],y[:10],columns
```

#### split the data into train and test 

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=24,stratify=y,test_size=0.3)
X_train.shape,X_test.shape
```

#### train the decision Tree

```python
dt_model=DecisionTreeClassifier()
```

```python
dt_model.fit(X_train,y_train)
```

## calculate the metrics for train and test

```python
y_pred_train=dt_model.predict(X_train)
y_pred_test=dt_model.predict(X_test)
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
print(classification_report(y_test,y_pred_test))
```

### plotting the decision tree

```python
from sklearn import tree
```

```python
tree.plot_tree(dt_model,feature_names=columns)
```

#### lets tune some hyper paramters so that there is no overfitting 

```python
dt_model_tuned=DecisionTreeClassifier(max_depth=3,min_samples_split=5)
```

```python
dt_model_tuned.fit(X_train,y_train)
```

```python
y_pred_train_tuned=dt_model_tuned.predict(X_train)
y_pred_test_tuned=dt_model_tuned.predict(X_test)
```

```python
print(classification_report(y_train,y_pred_train_tuned))
```

```python
print(classification_report(y_test,y_pred_test_tuned))
```

```python
tree.plot_tree(dt_model_tuned,feature_names=columns)
```
